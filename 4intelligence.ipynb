##Variáveis
Nesse momento, vamos olhar para cada variável e verificar se possuem valores nulos e quais são seus tipos.

```python
df.info()
```

Aqui podemos perceber que o formato de todas as variáveis estão corretos. A `data_tidy` está como o formate de data (datetime), `du` está com formato de número inteiro (int) e as demais variáveis estão corretamente como números decimais (float). 

Quanto aos valores ausentes, vemos que algumas variáveis apresentam valores nulos. 
* `renda_r` (variável independente)
* `massa_r` (variável independente)

Como a quantidade de valores ausentes dessas colunas é representativo ao nosso modelo, iremos preenchê-las com suas respectivas **medianas**. 

Além dessas duas, todas as variáveis dependentes (que são referentes ao consumo de energia nas 5 regiões) possuem valores ausentes. Isso quer dizer que as últimas entradas que estão vazias são justamente aquelas que precisamos projetar, de acordo com o nosso objetivo. Em outras palavras, **as últimas 22 entradas serão aquelas a serem projetadas pelo modelo de Machine Learning**. 



```python
mediana_renda_r = df.renda_r.median()
df.renda_r.fillna(mediana_renda_r, inplace = True)

mediana_massa_r = df.massa_r.median()
df.massa_r.fillna(mediana_massa_r, inplace = True)
```

```python
df.isnull().sum()
```

Agora podemos ver que os valores nulos das colunas `renda_r` e `massa_r` foram prenchidos por suas respectivas medianas. Assim conseguimos dar continuidade no nosso modelo, sem que haja problemas futuros (a biblioteca que usaremos lá na frente - scikit-learn - não permite usarmos valores nulos no dataframe)

Antes de seguirmos com o estudo, devemos excluir as últimas 22 entradas do dataframe (essas serão utilizadas para projetar a demanda de energia). 

Adicionalmente, iremos criar um novo dataframe `df_clean` desconsiderando a coluna de data `data_tidy`. Isso porque a biblioteca de Machine Learning que utilizaremos, não é possível utilizar variáveis com o tipo `datetime`. 

```python
#excluindo as últimas entradas

for i in range(206, 228):
  df = df.drop(i)
```

```python
#criando um novo dataframe sem a coluna de data

df_clean = df.drop(labels = 'data_tidy', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()

```

```python
X = df_clean.drop('ind_se', axis=1)
y = df_clean['ind_se']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)

```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```
