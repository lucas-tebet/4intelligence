Assim conseguimos dar continuidade no nosso modelo, sem que haja problemas futuros (a biblioteca que usaremos lá na frente - scikit-learn - não permite usarmos valores nulos no dataframe)

Adicionalmente, iremos criar um novo dataframe `df_clean` desconsiderando a coluna de data `data_tidy`. Isso porque a biblioteca de Machine Learning que utilizaremos, não é possível utilizar variáveis com o tipo `datetime`. 

```python
#criando um novo dataframe sem a coluna de data

df_clean = df.drop(labels = 'data_tidy', axis = 1)
```

#Modelo de Machine Learning
Agora que analisamos os nossos dados, fizemos as devidas alterações no dataset, estamos prontos para seguir com o nosso modelo de Machine Learning. 

Como esse estudo é refente a previsão de dados futuros,  trata-se de um aprendizado supervisionado. Onde, dadas informações de temperatura, produção industrial, comércio, renda, etc., a máquina aprenderá com valores passados (treino) e irá prever qual será o consumo de energia industrial na região Sudeste. 

Portanto, já sabemos que a nossa variável *target* é a coluna `ind_se` e as nossas variáveis independentes serão todas as demais. 

Por ser um aprendizado supervisionado, com mais de uma variável dependente, usaremos a **Regressão Linear Múltipla** no nosso modelo de Machine Learning. 

<p align=center>
<img src="https://stackabuse.s3.amazonaws.com/media/multiple-linear-regression-with-python-1.png" width="85%"></p>

Somente para ilustrar o modelo, acima está um exemplo de uma Regressão Linear Simples (com apenas uma variável independente). A diferença para a Múltipla é que existem mais de uma variável independente, o que _normalmente_ faz com que o modelo tenha uma melhor acurácia. No final, o objetivo é um mesmo: construir uma reta para definir os valores da variável dependente. 

##O modelo

```python
#importamos os pocotes que usaremos no modelo

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
#instanciamos o modelo

model = LinearRegression()

#dividimos entre dados de treino e de teste o dataframe já limpo, somente com as informações que queremos

X = df_clean.drop('ind_se', axis=1) #variáveis independentes
y = df_clean['ind_se'] #variável dependente

X_train, X_test, y_train, y_test = train_test_split(X, y)

#treinamos o nosso modelo com as variáveis de treino

model.fit(X_train, y_train)

#por fim, testamos 
y_pred = model.predict(X_test)
```

Após treinado o modelo, podemos ver todos os coeficientes que serão multiplicados pelas variáveis independentes e também o intercepto (posição em que a reta cruza o eixo Y). 

```python
print(model.coef_)
print(model.intercept_)
```

##Métricas
Para que possamos medir o resultado do nosso modelo, usaremos as métricas abaixo.

* **R²** usada para medir a acurácia do modelo. É um valor entre 0 e 1 e quando mais próximo de 1, maior a acurácia do modelo. 
* **MAE** é a média absoluta dos erros (Mean Absolute Error) que é representada pela soma dos erros de cada ponto referente a reta. 
* **MSE** é média da raíz quadrada dos erros (Mean Squared Error) que tende a penalizar exponencialmente os erros, tratando diferentemente erros pequenos e grandes. 

```python
print('R²: ', r2_score(y_test, y_pred))
print('MAE: ', mean_absolute_error(y_test, y_pred))
print('MSE: ', mean_squared_error(y_test, y_pred))
```
